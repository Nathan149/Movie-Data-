```python
df = df.dropna()
```

```python
#check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
#data types of columns

df.dtypes
```

```python
#change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
#Split the released column including year and location

df[['released', 'nation']] = df['released'].astype(str).str.split("(", expand = True)
```

```python
df.head()
```

```python
df = df.drop("nation", axis = 'columns')
```

```python
df.head()
```

```python
#create correct release year

df['release'] = df['released'].str[-5:-1]
```

```python
df.head()
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#display max rows

pd.set_option('display.max_rows', None)
```

```python
#correct negative values from gross column

df.loc[df['gross'] < 0, ['gross']] = df['gross'] * -1
        
```

```python
df_sorted = df.sort_values(by = ['gross'], inplace = False, ascending = False)
df_sorted.head()
```

```python
#checking for correleation
#creating a scatter plot for budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Gross earnings vs Budget')
plt.xlabel('Budget')
plt.ylabel('Gross earnings')

plt.show()
```

```python
#plot budget vs gross with seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"#088F8F"}, line_kws={"color":"blue"})
```

```python
df.corr(method='pearson') #pearson
```

```python
#correlation for numbered values in th dataframe

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation matrix for numeric values')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
#change string columns to numeric, to carry out correlation

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation matrix for numeric values')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```
